```r
# Read CSV and save in dataset
nba_final <- read.csv("nba_clean_good.csv")
```

```r
# Print out dataset
nba_final
```

```r
# Accessing all necessary data from dataset to make our MLR model

# First remove all other variables we're not using
new_data <- subset(as.data.frame(nba_final),
              select = c(GS, unclass(Pos1), Age, TOV, PTS, FGA))

# use this dataset for condition 2 plots to avoid changing new data
new_data_plots <- subset(as.data.frame(nba_final),
              select = c(GS, unclass(Pos1), Age, TOV, PTS, FGA))
# Now remove any row that contains a missing value
new_data <- new_data[complete.cases(new_data),]
dim(new_data)

# Print out the general summary for all nuemrical predictors
summary(new_data)

# Find the frequency of the categorical variable (Numerical Summary)
cat_freq = table(new_data$Pos1)
cat_freq

# Calculate standard deviations of each of the predictors (apart from categorical predictor)
sd_GS <- sd(new_data$GS)
sd_Age <- sd(new_data$Age)
sd_TOV <- sd(new_data$TOV)
sd_PTS <- sd(new_data$PTS)
sd_FGA <- sd(new_data$FGA)

# Print out our standard deviations
sd_GS
sd_Age
sd_TOV
sd_PTS
sd_FGA

write.csv(new_data, "cleaned_dataset")
```

```r
# Make the MLR model
og_model <- lm(new_data$PTS ~ new_data$Age + new_data$TOV + new_data$GS + new_data$FGA + new_data$Pos1)
og_model
```

```r
# Finding residual values
y_hat <- fitted(og_model)
e_hat <- resid(og_model)

# Plot Residual vs. Fitted
plot(x = y_hat, y = e_hat, main = "Residual vs Fitted", xlab = "Fitted",
     ylab = "Residuals")
```

```r
# Residuals v. Predictors grpahs for checking assumptions
# Plot each of the following Residual v. Predictors on a single grid


plot(x = new_data$GS, y = e_hat, main = "Residual vs GS",
     xlab = "Games Started (GS)", ylab = "Residual")
plot(x = new_data$Age, y = e_hat, main = "Residual vs Age",
     xlab = "Age", ylab = "Residual")
plot(x = new_data$FGA, y = e_hat, main = "Residual vs FGA",
     xlab = "Field Goals Attempted (FGA)", ylab = "Residual")
plot(x = new_data$TOV, y = e_hat, main = "Residual vs Turnovers",
     xlab = "Turnovers", ylab = "Residual")
```

```r
# Boxplot Graph for Categorical Predictor Variable
boxplot(e_hat ~ new_data$Pos1 , main = "Residual vs Primary Position",
        xlab = "Primary Position", ylab = "Residuals")
```

```r
# Check normality assumption by plotting the Normal QQ Plot
qqnorm(e_hat)
qqline(e_hat)
```

```r
#define problematic points
# useful values:
n <- nrow(new_data)
p <- length(coef(model_reduced))-1

# leverage cutoff
h_cut <- 2*((p+1)/n)

# cooks cutoff
D_cut <- qf(0.5, p+1, n-p-1) #median

# DFFITS cutoff
fits_cut <- 2*(sqrt((p+1)/n))

# DFBETAS cutoff
beta_cut <- 2*sqrt(n)
```

```r
# leverage
h_ii <- hatvalues(model_reduced)

# outlier
r_i <- rstandard(model_reduced)

# Cook's Distance
D_i <- cooks.distance(model_reduced)

# DFFITS
dffits_i <- dffits(model_reduced)

# DFBETAS
dfbetas_i <- dfbetas(model_reduced)
```

```r
# identify leverage points
which(h_ii>h_cut)

```

```r
# identify outliers
which(r_i > 4 | r_i < -4)

```

```r
# influential on all fitted values
which( D_i>D_cut)
```

```r

# influential on own fitted value
which(abs(dffits_i)> fits_cut )

```

```r
# influential on a coefficient
for(i in 1:8){
  print(paste0("Beta ", i-1))
  print(which(abs(dfbetas_i)> beta_cut))    # add your criteria here - this checks all betas in a loop
}
```
